```python
# import the necessary packages
import numpy as np
import argparse
import cv2
```

```python
#load the image
image = cv2.imread("data/cone1.jpeg")
#image = cv2.imread("data/image.jpg")
#image = cv2.imread("data/yellocone.jpg")
#image = cv2.imread("data/cone2.jpg")
#image = cv2.imread("data/coneDark.jpg")
```

```python
resized_image = cv2.resize(image,(200,200))
```

```python
hsv_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2HSV)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(hsv_image);
```

```python
resized_image = hsv_image
```

```python
plt.imshow(resized_image);
```

```python
# define the list of boundaries
boundaries = [
    ([110, 0, 60], [250, 126, 200]) #boundaries for purple color
]
```

```python
def colorPicker():
    return 1
```

```python
# loop over the boundaries
for (lower, upper) in boundaries:
    # create NumPy arrays from the boundaries
    lower = np.array(lower, dtype = "uint8")
    upper = np.array(upper, dtype = "uint8")

    # find the colors within the specified boundaries and apply
    # the mask
    mask = cv2.inRange(resized_image, lower, upper)
    output = cv2.bitwise_and(resized_image, resized_image, mask = mask)

    # show the images

    cv2.namedWindow('images',cv2.WINDOW_NORMAL)
    #cv2.createTrackbar('colorTracker','images', 2, 255, colorPicker)
    cv2.imshow("images", np.hstack([resized_image, output]))
    cv2.waitKey(0)
```
